```python
import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.callbacks import EarlyStopping

# Set fixed random seed for reproducibility
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# Suppress oneDNN optimization messages
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

# Load dataset
df = pd.read_csv(r'outliersremovedfile.csv')

# Split features and target
X = df.drop(columns=['Outcome'])  # Features
y = df['Outcome']  # Target variable

# Split dataset into 85% training, 15% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Normalize the features (Standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build Enhanced Neural Network Model
def build_model():
    model = keras.Sequential([
        keras.layers.Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
        keras.layers.BatchNormalization(),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dropout(0.3),
        keras.layers.BatchNormalization(),
        keras.layers.Dense(16, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

# Train and evaluate for 200, 400, and 800 epochs
for epochs in [200, 400, 800]:
    print(f"\n=== Training for {epochs} Epochs ===")
    model = build_model()

    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=32,
        validation_split=0.2,
        callbacks=[early_stopping],
        verbose=0
    )

    # Evaluate Model
    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    print(f"\nResults after {epochs} Epochs:")
    print(f"Training Accuracy: {train_acc*100:.2f}%")
    print(f"Testing Accuracy: {test_acc*100:.2f}%")
```

```python
import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Set fixed random seed for reproducibility
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# Suppress oneDNN optimization messages
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

# Load dataset
df = pd.read_csv(r'outliersremovedfile.csv')

# Split features and target
X = df.drop(columns=['Outcome'])  # Features
y = df['Outcome']  # Target variable

# Split dataset into 85% training, 15% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Normalize the features (Standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build Neural Network Model (3 Hidden Layers - matching paper architecture)
model = keras.Sequential([
    keras.layers.Dense(5, activation='relu', input_shape=(X_train.shape[1],)),  # Input layer (5 features)
    keras.layers.Dense(16, activation='relu'),  # First hidden layer (16 neurons)
    keras.layers.Dense(10, activation='relu'),  # Second hidden layer (10 neurons)
    keras.layers.Dense(5, activation='relu'),   # Third hidden layer (5 neurons)
    keras.layers.Dense(1, activation='sigmoid') # Output Layer (Binary Classification)
])

# Compile Model with SGD optimizer (as used in paper)
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train Model for 200, 400, and 800 epochs (as in paper)
for epochs in [200, 400, 800]:
    print(f"\nTraining 3-Hidden-Layer Model with {epochs} Epochs...")
    model.fit(X_train, y_train, epochs=epochs, batch_size=32, verbose=0)

    # Evaluate Model
    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    print(f"Epochs: {epochs}, Accuracy: {test_acc:.3f}, Training Accuracy: {train_acc*100:.2f}%, Testing Accuracy: {test_acc*100:.2f}%")
```

```python
import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Set fixed random seed for reproducibility
os.environ["PYTHONHASHSEED"] = "42"
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

# Load dataset
df = pd.read_csv('outliersremovedfile.csv')

# Split features and target
X = df.drop(columns=['Outcome'])
y = df['Outcome']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Loop for different epoch values
for epochs in [200, 400, 800]:
    print(f"\nTraining 2-Hidden-Layer Model with {epochs} Epochs...")

    # Reinitialize model inside loop to reset weights before retraining
    model = keras.Sequential([
        keras.layers.Dense(5, activation='relu', input_shape=(X_train.shape[1],)),
        keras.layers.Dense(26, activation='relu'),
        keras.layers.Dense(5, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    model.fit(X_train, y_train, epochs=epochs, batch_size=32, verbose=0)

    # Evaluate performance
    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"Epochs: {epochs}, Accuracy: {test_acc:.3f}, Training Accuracy: {train_acc*100:.2f}%, Testing Accuracy: {test_acc*100:.2f}%")

    # ROC Curve
    y_score = model.predict(X_test).ravel()
    y_true = y_test.values
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, color='green', linewidth=4, label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'r--', linewidth=2, label='Random guess')
    plt.title(f'ROC Curve for 2 Hidden Layer NN with {epochs} Epochs')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Data from the table
# K-fold cross-validation results
kfold_data = {
    'LR': {'TN': 114, 'FP': 41, 'FN': 119, 'TP': 425},
    'KNN': {'TN': 385, 'FP': 81, 'FN': 102, 'TP': 385},
    'SVM': {'TN': 417, 'FP': 49, 'FN': 114, 'TP': 417},
    'NB': {'TN': 142, 'FP': 76, 'FN': 91, 'TP': 390},
    'DT': {'TN': 118, 'FP': 101, 'FN': 115, 'TP': 365},
    'RF': {'TN': 137, 'FP': 74, 'FN': 96, 'TP': 392},
    'AB': {'TN': 139, 'FP': 69, 'FN': 94, 'TP': 397}
}

# Train/test splitting results
traintest_data = {
    'LR': {'TN': 37, 'FP': 11, 'FN': 32, 'TP': 130},
    'KNN': {'TN': 43, 'FP': 19, 'FN': 26, 'TP': 122},
    'SVM': {'TN': 42, 'FP': 13, 'FN': 27, 'TP': 128},
    'NB': {'TN': 43, 'FP': 25, 'FN': 26, 'TP': 116},
    'DT': {'TN': 45, 'FP': 44, 'FN': 24, 'TP': 97},
    'RF': {'TN': 45, 'FP': 21, 'FN': 24, 'TP': 120},
    'AB': {'TN': 45, 'FP': 17, 'FN': 24, 'TP': 124}
}

# Create subplots
fig, axes = plt.subplots(2, 7, figsize=(25, 10))
fig.suptitle('Confusion Matrices for DT, KNN, RF, NB, AB, LR, SVM Classifiers', y=1.02)

# Plot K-fold cross-validation results
for i, (model, data) in enumerate(kfold_data.items()):
    cm = np.array([
        [data['TN'], data['FP']],
        [data['FN'], data['TP']]
    ])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['0', '1'])
    disp.plot(ax=axes[0, i], cmap='Blues', values_format='d')
    axes[0, i].set_title(f'K-fold {model}')

# Plot Train/test splitting results
for i, (model, data) in enumerate(traintest_data.items()):
    cm = np.array([
        [data['TN'], data['FP']],
        [data['FN'], data['TP']]
    ])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['0', '1'])
    disp.plot(ax=axes[1, i], cmap='Blues', values_format='d')
    axes[1, i].set_title(f'Train/test {model}')

plt.tight_layout()
plt.show()
```
